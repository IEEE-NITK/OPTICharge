## 2. MRG 

N-step ahead control.

```python
def DFN_forward(F,x0,z0,Ir,beta,settings,control_settings):
    '''
        Given states, compute the forward dynamics.
    '''

    # Simulation horizon
    NT = 1;
    
    applied_action = beta*Ir
    
    
    x_sim = []
    z_sim = []
    
    x_sim.append(x0)
    z_sim.append(z0)
    
    # Simulate forward
    
    for k in range(NT):

        Fk = F(x0=x_sim[k], z0=z_sim[k], p=applied_action/p['Area'])


        x_tmp = Fk['xf'].full()
        z_tmp = Fk['zf'].full()
        v_tmp = Fk['qf'].full()/dt

        res = helper(x=x_tmp, z=z_tmp, p=[applied_action])

        SOCn = res['info_outs'][c_avgn_idx].full()
        c_avgn = res['info_outs'][c_avgn_idx].full()
        c_avgp = res['info_outs'][c_avgp_idx].full()
        cssn = res['info_outs'][cssn_idx].full()
        cssp = res['info_outs'][cssp_idx ].full()
        c_exx = res['info_outs'][cex_idx].full()
        etan = res['info_outs'][etan_idx].full()
        etap = res['info_outs'][etap_idx].full()
        ce0n = res['info_outs'][ce0n_idx].full()
        ce0p = res['info_outs'][ce0p_idx].full()
        etasLn = res['info_outs'][etasLn_idx].full()[0]
        Volt = res['info_outs'][Volt_idx].full()
        nLis = res['info_outs'][nLis_idx].full()
        nLie = res['info_outs'][nLie_idx].full()
        i0n = res['info_outs'][i0n_idx].full()
        i0p = res['info_outs'][i0p_idx].full()

        c_s_n = x_tmp[out_csn_idx]
        c_s_p = x_tmp[out_csp_idx]
        c_ex = x_tmp[out_ce_idx]
        Temp = x_tmp[out_T_idx]

        out_phisn = z_tmp[out_phisn_idx]
        out_phisp = z_tmp[out_phisp_idx]
        out_ien = z_tmp[out_ien_idx]
        out_iep = z_tmp[out_iep_idx]
        out_phie = z_tmp[out_phie_idx]
        out_jn = z_tmp[out_jn_idx]
        out_jp = z_tmp[out_jp_idx]

        SOCn = SOCn/p['c_s_n_max']
        SOCn = float(sum(SOCn)/len(SOCn))
        c_ss_n = res['info_outs'][cssn_idx]
    
    # Compute constraint violation
#     ipdb.set_trace()
    control_settings['constraints']['temperature']['max'] = 273+40
    yc1 = control_settings['constraints']['etasLn']['min'] - etasLn
    yc1 = yc1[0] # make sure the value is not array type.
    yc2 = Temp - control_settings['constraints']['temperature']['max']
    yc = np.array([yc1,yc2])
#     yc3 = np.max(cssn/p['c_s_n_max']) - control_settings['constraints']['normalized_cssn']['max']    
#     yc = np.array([yc1,yc2,yc3])

    
    
    return yc
```

```python
def bisection_DFN(F,x0,z0,Ir,settings,control_settings):
    
    maxIters = 100
    beta = np.zeros(maxIters)
    beta[0] = 1
    
    beta_low = 0
    beta_high = 1
    beta_err = 0.005
    
    ## Bisection algorithm
    
    for idx in range(maxIters):

        # Simulate DFN model forward
        yc = DFN_forward(F,x0,z0,Ir,beta[idx],settings,control_settings)

        # constraint is satisfied
        if all( yc <= 0):
            if idx == 0:
                beta_star = 1
                break
            elif ( abs(beta[idx] - beta[idx-1]) <= beta_err ):
                beta_star = beta[idx]
                break
            else:
                beta_low = beta[idx]
            
        # constraint is violated
        else:
            beta_high = beta[idx]
            
        # Bisection
        beta[idx+1] = (beta_high + beta_low) / 2
#     ipdb.set_trace()    
    print('Iter = {} | beta = {:4f}'.format(idx,beta[idx]))
    beta_star = beta[idx]
    
    return beta_star
```

```python
# MAIN

init_V = 3.5
init_T = 300

ACTION_VEC=[]
SOC_VEC=[]
T_VEC=[]
VOLTAGE_VEC=[]
RETURN_VALUE=0
TIME_VEC = []
done=False


x_sim = []
z_sim = []
v_sim = []

x_sim.append(x_init)
z_sim.append(z_init)
v_sim.append(init_V)


cex_sim = []
etasLn_sim = []
cssn_sim = []
cssp_sim = []
cavgn_sim = []
cavgp_sim = []
nLis_sim = []
phi_s_n_sim = []
phi_s_p_sim = []
ien_sim = []
iep_sim = []
jn_sim = []
jp_sim = []
phi_e_sim = []


beta_sim = []

tt=0
TIME_VEC.append(tt)


## MRG Parameters
Max_chg = -130
Iref = Max_chg


for idx in range(10000):
    

    # Reference Governor
    beta = bisection_DFN(F, x_sim[idx], z_sim[idx], Iref, settings, control_settings)
    
    # Applied action
    applied_action = beta * Iref
    
    # Step-forward
    Fk = F(x0=x_sim[idx], z0=z_sim[idx], p=applied_action/p['Area'])
    
    x_nxt = Fk['xf'].full()
    z_nxt = Fk['zf'].full()
    v_nxt = Fk['qf'].full()/dt

    res = helper(x=x_nxt, z=z_nxt, p=[applied_action/p['Area']])
    
    # save info.
    x_sim.append(x_nxt)
    z_sim.append(z_nxt)
    v_sim.append(v_nxt)    
    beta_sim.append(beta)
    
    # Extract info.
    SOCn = res['info_outs'][c_avgn_idx].full()
    c_avgn = res['info_outs'][c_avgn_idx].full()
    c_avgp = res['info_outs'][c_avgp_idx].full()
    cssn = res['info_outs'][cssn_idx].full()
    cssp = res['info_outs'][cssp_idx ].full()
    c_exx = res['info_outs'][cex_idx].full()
    etan = res['info_outs'][etan_idx].full()
    etap = res['info_outs'][etap_idx].full()
    ce0n = res['info_outs'][ce0n_idx].full()
    ce0p = res['info_outs'][ce0p_idx].full()
    etasLn = res['info_outs'][etasLn_idx].full()[0]
    Volt = res['info_outs'][Volt_idx].full()
    nLis = res['info_outs'][nLis_idx].full()
    nLie = res['info_outs'][nLie_idx].full()
    i0n = res['info_outs'][i0n_idx].full()
    i0p = res['info_outs'][i0p_idx].full()

    c_s_n = x_nxt[out_csn_idx]
    c_s_p = x_nxt[out_csp_idx]
    c_ex = x_nxt[out_ce_idx]
    Temp = x_nxt[out_T_idx]

    out_phisn = z_nxt[out_phisn_idx]
    out_phisp = z_nxt[out_phisp_idx]
    out_ien = z_nxt[out_ien_idx]
    out_iep = z_nxt[out_iep_idx]
    out_phie = z_nxt[out_phie_idx]
    out_jn = z_nxt[out_jn_idx]
    out_jp = z_nxt[out_jp_idx]

    SOCn = SOCn/p['c_s_n_max']
    SOCn = float(sum(SOCn)/len(SOCn))
    c_ss_n = res['info_outs'][cssn_idx]
    
    # Save essential vector
    tt += dt
    TIME_VEC.append(tt)
    ACTION_VEC.append(applied_action)
    SOC_VEC.append(SOCn)
    T_VEC.append(Temp)
    VOLTAGE_VEC.append(v_nxt[0][0])
    
    # Save simulation info.
    cex_sim.append(c_ex)
    etasLn_sim.append(etasLn)
    cssn_sim.append(cssn)
    cssp_sim.append(cssp)
    cavgn_sim.append(c_avgn)
    cavgp_sim.append(c_avgp)
    nLis_sim.append(nLis)
    
    phi_s_n_sim.append(out_phisn)
    phi_s_p_sim.append(out_phisp)
    ien_sim.append(out_ien)
    iep_sim.append(out_iep)
    jn_sim.append(out_jn)
    jp_sim.append(out_jp)
    phi_e_sim.append(out_phie)
    
    print('SOC: {:.2f}'.format(SOCn))
    if SOCn >= 0.8:
        break

```
